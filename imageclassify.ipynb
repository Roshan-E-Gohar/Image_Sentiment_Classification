```python

# Remove the directory containing images of the class you want to exclude
class_to_remove = '.ipynb_checkpoints'  # Replace 'class_to_remove' with the name of the class you want to remove
dataset_dir ='C:\\Users\\BEST BUY\\Image_Sentiment_Classification\\data'  # Replace 'data' with the path to your dataset directory
class_dir_to_remove = os.path.join(dataset_dir, class_to_remove)

if os.path.exists(class_dir_to_remove):
    # Remove the directory
    os.rmdir(class_dir_to_remove)
    print(f"Removed directory '{class_to_remove}'")
else:
    print(f"Directory '{class_to_remove}' not found")

# Recreate the dataset without the removed class
data = tf.keras.utils.image_dataset_from_directory(dataset_dir)
```
